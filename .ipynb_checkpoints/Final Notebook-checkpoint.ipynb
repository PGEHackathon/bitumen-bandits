To determine which features from the "HackathonData2024.csv" are correlated with the "Avg Pump Difference" feature, I will employ a variety of correlation techniques. These include:

- **Pearson Correlation:** Measures linear correlation between two variables. It assumes a normal distribution of the attributes.

- **Spearman's Rank Correlation:** Non-parametric measure of rank correlation. It assesses how well the relationship between two variables can be described using a monotonic function.

- **Kendall's Tau:** Another non-parametric measure that assesses the strength and direction of association between two variables.

```python
import pandas as pd

# Load the HackathonData2024.csv file
file_path = 'HackathonData2024.csv'
hackathon_data = pd.read_csv(file_path)
```

The **"HackathonData2024.csv"** file contains multiple features along with the **"Avg Pump Difference"** which is our target variable. To find the correlations, I will first preprocess the data. This includes:

1. **Handling Missing Values:** Techniques like imputation or removal of rows/columns with missing values will be considered.

2. **Encoding Categorical Data:** Transforming non-numeric columns into a format that can be used in correlation calculations.

After preprocessing, I'll calculate Pearson, Spearman's, and Kendall's Tau correlations for each feature with the **"Avg Pump Difference"**.

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Preprocessing the data
# Handling missing values by imputing with the median (for numerical features)
numerical_features = hackathon_data.select_dtypes(include=[np.number]).columns
hackathon_data[numerical_features] = hackathon_data[numerical_features].apply(lambda x: x.fillna(x.median()),axis=0)

# Encoding categorical data
categorical_features = hackathon_data.select_dtypes(include=['object']).columns
label_encoders = {col: LabelEncoder() for col in categorical_features}

for col in categorical_features:
    hackathon_data[col] = label_encoders[col].fit_transform(hackathon_data[col])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculating correlation coefficients
pearson_corr = hackathon_data.corr(method='pearson')['Avg Pump Difference']
spearman_corr = hackathon_data.corr(method='spearman')['Avg Pump Difference']
kendall_corr = hackathon_data.corr(method='kendall')['Avg Pump Difference']

# Creating a dataframe to display correlations
correlation_df = pd.DataFrame({
    'Pearson': pearson_corr,
    'Spearman': spearman_corr,
    'Kendall': kendall_corr
})

# Dropping the 'Avg Pump Difference' correlation with itself
correlation_df = correlation_df.drop('Avg Pump Difference')

# Displaying the correlations in a more visual way
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Avg Pump Difference')
plt.show()

#correlation_df.sort_values(by='Pearson', ascending=False)
```

1. **Strong Negative Correlations:**

2. **Moderate Positive Correlations:**
   - The **"Fluid System"** feature shows a moderate positive correlation. This suggests that changes in the fluid system could significantly impact the average pump difference.

**Next Steps:**

1. **Model Tuning:**
   - Fine-tuning the Random Forest model to optimize its performance.

2. **Uncertainty Estimation:**
   - Implementing a method to generate 100 realizations for each prediction to model the uncertainty.

3. **Final Predictions:**
   - Applying the tuned model to predict the **"Avg Pump Difference"** for the 15 wells in your dataset, along with the uncertainty estimates.

```python
from sklearn.utils import resample

# Number of bootstrap samples
n_bootstrap_samples = 100

# Lists to store predictions for each bootstrap sample
bootstrap_predictions = []

# Performing bootstrapping
for i in range(n_bootstrap_samples):
    # Creating a bootstrap sample
    X_bootstrap, y_bootstrap = resample(X, y, random_state=i)
    
    # Training the Random Forest model on the bootstrap sample
    rf_bootstrap = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42)
    rf_bootstrap.fit(X_bootstrap, y_bootstrap)
    
    # Making predictions on the test set and storing them
    y_pred_bootstrap = rf_bootstrap.predict(X_test)
    bootstrap_predictions.append(y_pred_bootstrap)

# Converting the list of predictions to a NumPy array for easy computation
bootstrap_predictions = np.array(bootstrap_predictions)

# Calculating the average prediction and the uncertainty (standard deviation) for each well
average_prediction = np.mean(bootstrap_predictions, axis=0)
uncertainty = np.std(bootstrap_predictions, axis=0)

# Displaying the results for a few wells as an example
average_prediction[:5], uncertainty[:5]
```

```python
# Assuming the first 15 wells in the test set are the wells of interest
# (In practice, the specific wells of interest should be identified)

# Selecting the first 15 wells from the test set
X_wells_of_interest = X_test.iloc[:15]

# Lists to store predictions for each well for each bootstrap sample
predictions_wells_of_interest = {i: [] for i in range(1, 16)}  # Dictionary to store predictions for each well

# Making predictions for each well using each bootstrap model
for bootstrap_model_predictions in bootstrap_predictions:
    for i, well_prediction in enumerate(bootstrap_model_predictions[:15], start=1):
        predictions_wells_of_interest[i].append(well_prediction)

# Converting the dictionary of lists into a dataframe format suitable for the solution.csv file
solution_df = pd.DataFrame(predictions_wells_of_interest).transpose()
solution_df.columns = [f'R_{i+1}' for i in range(100)]  # Naming columns as R_1, R_2, ..., R_100

# Adding a column for average pump difference estimation (mean of the 100 realizations)
solution_df['Est Pump Difference'] = solution_df.mean(axis=1)

# Reordering columns to match the format of solution.csv
solution_df = solution_df[['Est Pump Difference'] + [f'R_{i+1}' for i in range(100)]]

solution_df.head()  # Displaying the first few rows of the solution dataframe
```
